# Seaborn – Class 2: Categorical Plots


## What Are Categorical Plots?

Categorical plots visualize data where at least one variable is **categorical**  
(e.g., day of week, gender, class).

Seaborn gives several easy options to:
- Compare **counts** or **averages** across categories
- Understand spread & distribution


```python
import seaborn as sns
import matplotlib.pyplot as plt

tips  = sns.load_dataset("tips")
```

```python
tips
```

## barplot() vs countplot()

```python
sns.barplot(x='day',y='total_bill',data=tips)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.countplot(x='day',data=tips)
```

```python
sns.countplot(x='sex',data=tips)
#mostly use in machine learning for class Imabalance
```

## boxplot()

```python
sns.boxplot(x='sex',y='total_bill',data=tips)
```

## stripplot() & swarmplot()

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

| Plot        | Best for |
|-------------|-----------|
| **barplot** | Compare average values across categories |
| **countplot** | Show how many items fall in each category |
| **boxplot** | See median, quartiles, and outliers |
| **violinplot** | Same as boxplot + shows distribution |
| **stripplot** | View every observation (good for small data) |
| **swarmplot** | Like stripplot but avoids overlap |


```python
sns.countplot(x='sex',hue='smoker',data=tips)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.barplot(x='day',y='tip',data=tips,hue='time')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python

sns.boxplot(x='smoker',y='tip',data=tips,hue='sex')
```
